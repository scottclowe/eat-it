```python
import sklearn
import pandas as pd
import numpy as np
import datetime
```

```python
from IPython.display import HTML
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import scipy
```

```python
from eat_it import params
```

```python
train = pd.read_csv('data/train.csv', encoding="utf-8")
# Add age in days
end_dt = datetime.datetime.strptime('2015-1-1', "%Y-%m-%d")
train['Age'] = [(end_dt - datetime.datetime.strptime(open_dt, "%m/%d/%Y")).days for open_dt in train['Open Date']]
# add size as boolean
train['isBig'] = train['City Group']=='Big Cities'
# add boolean for type
train['isIL'] = train['Type']=='IL'
```

```python
gtest = pd.read_csv('data/genuinetest.csv', encoding="utf-8")
# Add age in days
end_dt = datetime.datetime.strptime('2015-1-1', "%Y-%m-%d")
gtest['Age'] = [(end_dt - datetime.datetime.strptime(open_dt, "%m/%d/%Y")).days for open_dt in gtest['Open Date']]
# add size as boolean
gtest['isBig'] = gtest['City Group']=='Big Cities'
# add boolean for type
gtest['isIL'] = gtest['Type']=='IL'
```

```python
data = pd.concat((train,gtest))
```

```python
data
```

```python
data['missingSource'] = data[params.xor_cols].apply(lambda x: np.all(x==0), axis=1)
```

```python
misSource = data['missingSource'].values
```

```python
# List all the columns we can train on
cols = ['P'+str(i) for i in range(1,38)]

cols.append('Age')
cols.append('isBig')
cols.append('isIL')
cols.append('missingSource')
cols.append('revenue')

data_mtx = data.as_matrix(columns=cols).astype(np.float)
data2_mtx = data[np.logical_not(misSource)].as_matrix(columns=cols).astype(np.float)
```

```python
rho,p = scipy.stats.spearmanr(data_mtx)
sns.heatmap(rho)
```

```python
rho,p = scipy.stats.spearmanr(data2_mtx)
sns.heatmap(rho)
```

```python
# Reorder indices based on previous manual clustering
index = [i-1 for l in params.manual_clusters for i in l]
rho,p = scipy.stats.spearmanr(data_mtx[:,index])
sns.heatmap(rho)
plt.show()
rho,p = scipy.stats.spearmanr(data2_mtx[:,index])
sns.heatmap(rho)
plt.show()
```
