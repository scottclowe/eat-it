```python
sublog = pd.read_csv('../submission_log.tsv', delimiter='\t')
```

```python
all_uids = np.union1d(sublog['uid1'].values, sublog['uid2'].values)
print(all_uids)
```

```python
# Random state, seeded from entropy or clock
rng = np.random.RandomState(None)
```

```python
# Random state, seeded from entropy or clock
while True:
    idx = rng.randint(len(gtest))
    uid = gtest.iloc[idx]['Id']
    if not np.any(all_uids==str(uid)):
        break
    print('Failed for {}, {}'.format(idx,uid))

print()
print('Succeeded for {}, {}'.format(idx,uid))
print()

print(gtest[idx:(idx+1)][['Age','Type','Id']])
print(repr(gtm[int(gtest[idx:(idx+1)][['Id']].values)]))
```
